```python
import cv2
from matplotlib import pyplot as plt
```

```python
#翻轉：實作上下左右的翻轉
```

```python
img_rgb=cv2.imread('dog.jpg',cv2.IMREAD_COLOR)
```

```python
img_rgb=img_rgb[::-1,:,:]
```

```python
cv2.imshow('dog_rgb',img_rgb)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.imshow(img_rgb)
plt.title('my picture')
plt.show()
```

```python
#縮放：比較鄰近差值與雙立方插值 (或雙線性插值) 的圖片品質
```

```python
img_rgb=cv2.imread('dog.jpg',cv2.IMREAD_COLOR)
```

```python
plt.imshow(img_rgb)
plt.title('my picture')
plt.show()
```

```python
img_rgb=cv2.resize(img_rgb,(400,400),cv2.INTER_NEAREST)
plt.imshow(img_rgb)
plt.title('my picture')
plt.show()
```

```python
img_rgb=cv2.resize(img_rgb,(400,400),cv2.INTER_LINEAR)
plt.imshow(img_rgb)
plt.title('my picture')
plt.show()
```

```python
#平移：建立 Translation Transformation Matrix 來做平移
```

```python
import numpy as np
```

```python
H = np.float32([[1,0,100],[0,1,50]])
rows,cols = img_rgb.shape[:2]
res = cv2.warpAffine(img_rgb,H,(rows,cols)) 
plt.imshow(res)
plt.title('my picture')
plt.show()
```
